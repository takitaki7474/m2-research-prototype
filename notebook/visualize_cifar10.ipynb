```python
import numpy as np
import matplotlib.pyplot as plt
import chainer
```

#### cifar10のクラスラベル
- 0: airplane (飛行機)
- 1: automobile (自動車)
- 2: bird (鳥)
- 3: cat (猫)
- 4: deer (鹿)
- 5: dog (犬)
- 6: frog (カエル)
- 7: horse (馬)
- 8: ship (船)
- 9: truck (トラック)

```python
(train, test) = chainer.datasets.get_cifar10()
print("訓練データ数: ", len(train))
print("テストデータ数: ", len(test))
```

```python
# 複数画像の表示
fig,ax = plt.subplots(nrows=3,ncols=5,sharex=True,sharey=True,figsize=(18,10))
ax = ax.flatten()
for i in range(15):
    img = np.dstack((train[i][0][0],train[i][0][1],train[i][0][2]))
    ax[i].imshow(img,interpolation='none')

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
# 1枚画像の表示
plt.figure(figsize=(4,4))
img = np.dstack((train[0][0][0], train[0][0][1], train[0][0][2]))
plt.imshow(img)
plt.show()
```

```python
# 特定クラスラベルの画像を表示
def display_specific_img(class_label):
    specific_img = get_specific_img(class_label)
    fig,ax = plt.subplots(nrows=3,ncols=5,sharex=True,sharey=True,figsize=(18,10))
    ax = ax.flatten()
    for i in range(15):
        img = specific_img[i][0]
        img = np.dstack((img[0], img[1], img[2]))
        ax[i].imshow(img,interpolation='none')

    ax[0].set_xticks([])
    ax[0].set_yticks([])
    plt.tight_layout()
    plt.show()

# 特定クラスラベル画像の取得
def get_specific_img(class_label):
    specific_img = []
    for data in train:
        label = data[1]
        if label == class_label:
            specific_img.append(data)
    return specific_img    
```

```python
display_specific_img(class_label=0)
```
