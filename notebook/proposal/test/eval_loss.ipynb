```python
def eval_err_speed(losses, a=50, n=10):
    err_speed = 0.0
    E = losses
    maxepoch = len(losses)
    ks = maxepoch//n
    for k in range(1, ks+1):
        err_speed += (-a / k) * (E[n*k-1] - E[n*k - n])
    return err_speed
```

```python
# 訓練誤差の推移の基準関数としてy = 0.8^x + 0.25を配列で取得
def get_standard_losses(epoch=80):
    x = list(range(1, epoch+1))
    standard_losses = []
    for i in x:
        standard_losses.append(0.8**i + 0.25)
    return standard_losses
```

```python
import matplotlib.pyplot as plt
# lossのplotのstub
def plot_loss(train_losses):
    max_axisX = len(train_losses)
    plt.figure(figsize=(7,5))
    plt.plot(range(max_axisX), train_losses)
    plt.ylim(bottom=0)
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend(['train/loss'])
    plt.grid()
    plt.show()
```

```python
train_losses, test_losses = losses_stub()
err_speed = eval_err_speed(train_losses)
print("train error speed: {0}".format(err_speed))
plot_loss(train_losses)
```
