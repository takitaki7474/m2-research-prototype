```python
import matplotlib.pyplot as plt
import os
import json
```

```python
def load_err_speeds(eval_name, result_versions):
    train_err_speeds = []
    test_err_speeds = []
    for v in result_versions:
        path = os.path.join("./result", eval_name, v, "err_speed.json")
        with open(path) as f:
            log = json.load(f)
        train_err_speeds.append(log["train_err_speed"])
        test_err_speeds.append(log["test_err_speed"])
    return train_err_speeds, test_err_speeds
```

```python
result_versions = ["v1", "v2", "v3", "v4", "v5", "v6", "v7", "v8", "v9", "v10"]
train_err_speeds1, test_err_speeds1 = load_err_speeds("eval1", result_versions)
train_err_speeds2, test_err_speeds2 = load_err_speeds("eval2", result_versions)
```

```python
print(train_err_speeds1)
print("max: {0}".format(max(train_err_speeds1)))
print("min: {0}\n".format(min(train_err_speeds1)))
print(train_err_speeds2)
print("max: {0}".format(max(train_err_speeds2)))
print("min: {0}\n".format(min(train_err_speeds2)))
print(test_err_speeds1)
print("max: {0}".format(max(test_err_speeds1)))
print("min: {0}\n".format(min(test_err_speeds1)))
print(test_err_speeds2)
print("max: {0}".format(max(test_err_speeds2)))
print("min: {0}".format(min(test_err_speeds2)))
```

```python
# 訓練誤差収束速度の比較
plt.figure(figsize=(12,10))
plt.ylim(0, 100)
plt.xticks(range(1,11), range(1,11))
plt.plot(range(1,11), train_err_speeds1,  color="blue", label="with query update")
plt.plot(range(1,11), train_err_speeds2, color="red", label="without query update")
plt.legend(fontsize=18)
plt.grid(True)
```

```python
# 汎化誤差収束速度の比較
plt.figure(figsize=(12,10))
plt.ylim(-700, 40)
plt.xticks(range(1,11), range(1,11))
plt.plot(range(1,11), test_err_speeds1,  color="blue", label="with query update")
plt.plot(range(1,11), test_err_speeds2, color="red", label="without query update")
plt.legend(fontsize=18)
plt.grid(True)
```
