```python
import matplotlib.pyplot as plt
import os
import json
```

```python
def load_err_speeds(result_versions):
    train_err_speeds = []
    test_err_speeds = []
    for v in result_versions:
        path = os.path.join("./result/eval2", v, "err_speed.json")
        with open(path) as f:
            log = json.load(f)
        train_err_speeds.append(log["train_err_speed"])
        test_err_speeds.append(log["test_err_speed"])
    return train_err_speeds, test_err_speeds
```

```python
result_versions = ["v1", "v2", "v3", "v4", "v5", "v6", "v7", "v8"]
train_err_speeds, test_err_speeds = load_err_speeds(result_versions)
```

```python
plt.figure(figsize=(7,5))
plt.xticks(range(1,9), range(1,9))
plt.ylim(0.0, 40.0)
plt.plot(range(1,9), train_err_speeds, marker="o", color="blue")
plt.plot(range(1,9), test_err_speeds, marker="o", color="red")
```
