```python
import matplotlib.pyplot as plt
import pandas as pd
import json
import os
```

```python
# 従来方法の評価対象logの読み込み
eval_list = ["eval4", "eval5", "eval6", "eval7", "eval8", "eval12", "eval13", "eval14", "eval15", "eval16"]
model_list = ["v" + str(i) for i in range(1, 21)]
conventional_accs = []

for eval_v in eval_list:
    accs = []
    for model_v in model_list:
        path = os.path.join("./result", eval_v, model_v, "log.json")
        with open(path) as f:
            log = json.load(f)
        acc = log[-1]["test_acc"]
        accs.append(acc)
    conventional_accs.append(accs)
```

```python
# 従来方法の精度推移の可視化
xlist = list(range(1,21))

fig1 = plt.figure(figsize=(13,8))
plt.ylim(0.38, 0.92)
plt.xticks(xlist, list(range(120, 120*20+1, 120)))
plt.plot(xlist, conventional_accs[0], color="tab:blue")
plt.plot(xlist, conventional_accs[1], color="tab:orange")
plt.plot(xlist, conventional_accs[2], color="tab:green")
plt.plot(xlist, conventional_accs[3], color="tab:red")
plt.plot(xlist, conventional_accs[4], color="tab:purple")
plt.plot(xlist, conventional_accs[5], color="tab:brown")
plt.plot(xlist, conventional_accs[6], color="tab:pink")
plt.plot(xlist, conventional_accs[7], color="tab:gray")
plt.plot(xlist, conventional_accs[8], color="tab:olive")
plt.plot(xlist, conventional_accs[9], color="tab:cyan")
plt.grid(True)
```
