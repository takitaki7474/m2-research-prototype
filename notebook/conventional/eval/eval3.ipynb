```python
import matplotlib.pyplot as plt
import os
import json
```

```python
def load_test_losses(version):
    test_losses = []
    path = os.path.join("./result/eval3", version, "log.json")
    with open(path) as f:
        logs = json.load(f)
    for log in logs:
        test_losses.append(log["test_loss"])
    return test_losses
```

```python
version = "v10"
v10_test_losses = load_test_losses(version)
version = "extra_v10"
extra_v10_test_losses = load_test_losses(version)
```

```python
plt.figure(figsize=(12,10))
plt.ylim(0.0, 1.2)
plt.plot(range(0,100), v10_test_losses,  color="blue", label="test loss 1")
plt.plot(range(0,100), extra_v10_test_losses, color="red", label="test loss 2")
plt.legend(fontsize=18)
plt.grid(True)
```
