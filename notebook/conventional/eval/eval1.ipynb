```python
import matplotlib.pyplot as plt
import os
import json
```

```python
def plot_accs_transition(dataNs, accs, ylim1=0.75, ylim2=1.0, color="blue"):
    plt.figure(figsize=(7,5))
    plt.xticks(dataNs, dataNs)
    plt.ylim(ylim1, ylim2)
    plt.plot(dataNs, accs, marker="o", color=color)
```

```python
def load_accs(result_versions):
    log_paths = []
    for v in result_versions:
        path = os.path.join("./result/eval1", v, "log.json")
        log_paths.append(path)
    accs = []
    for path in log_paths:
        with open(path) as f:
            log = json.load(f)
        accs.append(log[-1]["test_acc"])
    return accs
```

### v9-v16

```python
result_versions = ["v9", "v10", "v11", "v12", "v13", "v14", "v15", "v16"]
accs = load_accs(result_versions)
print(list(accs))
```

```python
dataNs = list(range(900, 7201, 900))
plot_accs_transition(dataNs, accs)
```

```python
result_versions = ["v1", "v2", "v3", "v4", "v5", "v6", "v7"]
accs = load_accs(result_versions)
print(list(accs))
```
